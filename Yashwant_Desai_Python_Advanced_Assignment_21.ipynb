# Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?

A probability distribution is a mathematical function that describes the likelihood of different outcomes or values occurring in a random experiment or process. It provides a way to model and quantify uncertainty, even though individual outcomes may be random. Probability distributions can be discrete (with countable, distinct values) or continuous (with a range of possible values), and they are used in statistics and probability theory to make probabilistic predictions and analyze random processes. In Python, libraries like NumPy and SciPy are commonly used to work with probability distributions for simulations and statistical analyses.

Predicting values in the context of probability distributions and randomness, we are not making deterministic predictions for individual outcomes. Instead, we are making probabilistic predictions about the likelihood of various outcomes occurring. Probability distributions provide a way to describe the patterns of randomness and quantify the chances of different outcomes.

```python
import numpy as np

num_simulations = 1000
coin_flips = np.random.choice(['Heads', 'Tails'], num_simulations, p=[0.5, 0.5])
empirical_probability_heads = np.mean(coin_flips == 'Heads')

print("Empirical Probability of Getting Heads:", empirical_probability_heads)
```

# Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?

Yes, there is a distinction between true random numbers and pseudo-random numbers.

The distinction between true random numbers and pseudo-random numbers lies in their sources and predictability:

True random numbers come from unpredictable physical processes and offer genuine randomness.
Pseudo-random numbers are generated by algorithms and mimic randomness but are fully predictable if you know the seed.
Pseudo-random numbers are widely used and considered "good enough" for most non-cryptographic applications due to their reproducibility, efficiency, and reasonable approximation of randomness. Python's random module provides pseudo-random number generation. For security-critical applications, true random numbers are essential. Python's secrets module offers cryptographically secure random numbers. The choice between the two depends on the specific requirements of the application.

# Q3. What are the two main factors that influence the behaviour of a "normal" probability distribution?

The behavior of a normal probability distribution is primarily influenced by two main factors i.e. mean and standard deviation.

Mean (μ or mu): The mean represents the central or average value of the distribution and determines the location of the peak of the bell curve. It is the measure of central tendency for the distribution.

Standard Deviation (σ or sigma): The standard deviation measures the spread or dispersion of the values in the distribution. A smaller standard deviation results in a narrower, taller curve, while a larger standard deviation leads to a wider, shorter curve. It quantifies the variability in the distribution.


# Q4. Provide a real-life example of a normal distribution.

A common real-life example of a normal distribution is the distribution of human heights in a population. In most populations, human heights tend to follow a normal distribution. Below is an example.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mean_height = 170  
std_deviation = 7  
height_data = np.random.normal(mean_height, std_deviation, 10000)
plt.hist(height_data, bins=30, density=True, alpha=0.6, color='b')
x = np.linspace(mean_height - 3 * std_deviation, mean_height + 3 * std_deviation, 100)
pdf = norm.pdf(x, mean_height, std_deviation)
plt.plot(x, pdf, 'r', linewidth=2)
plt.title('Distribution of Human Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.show()
```

# Q5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows?

The behavior of a probability distribution in the short term is characterized by variability, inaccuracy, and randomness, especially with a small number of trials or samples. The observed probabilities may deviate significantly from the expected values.

However, as the number of trials or samples grows, the distribution tends to exhibit more predictable and stable behavior. It converges toward its theoretical or expected values, becoming more stable and consistent. This is known as the law of large numbers, and it illustrates how the distribution's behavior becomes more accurate and less random with an increasing number of trials or samples.

# Q8. What is the relationship between exponentiation and logarithms?

Exponentiation and logarithms are mathematical operations that are closely related and inverse of each other. Exponentiation involves raising a base number to a certain power, while logarithms tell you what exponent is needed to obtain a specific result when a base is raised to that exponent. They are fundamental in mathematics, science, and engineering for solving equations involving exponential growth or decay and understanding various natural processes. Python provides the ** operator for exponentiation and the math.log() function for logarithmic calculations.

```python
import math

base = 2
exponent = 3
result = base ** exponent
print(result)
```

```python
import math

base = 2
result = 8
exponent = math.log(result, base)
print(exponent)
```

# Q9. What are the three logarithmic functions that Python supports?

Python's math module supports three main logarithmic functions:

o Natural Logarithm (Base e): math.log(x) computes the natural logarithm of a number x with the base e (Euler's number, approximately 2.71828).

o Common Logarithm (Base 10): math.log10(x) calculates the base-10 logarithm of a number x.

o Custom Base Logarithm: You can compute a logarithm with a custom base b by using the change of base formula, which involves the math.log() function. This allows you to calculate logarithms with bases other than e or 10.

Below are examples

```python
import math
result = math.log(10)
print(result)
```

```python
import math
result = math.log10(100)
print(result)
```

```python
import math
x = 100
base = 2
result = math.log(x) / math.log(base)
print(result)
```
